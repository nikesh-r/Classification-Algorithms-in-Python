```
#importing the libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```
from google.colab import files
uploaded = files.upload()
```

```
import io
dataset = pd.read_csv(io.BytesIO(uploaded['KDDTrain+.csv']))
dataset.head(10)
```

```
dataset.isnull().values.any()
```

```
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

featured_columns = dataset[['protocol_type', 'service', 'flag', 'count', 'srv_count']]

featured_columns['protocol_type'] = label_encoder.fit_transform(featured_columns['protocol_type'])
featured_columns['service'] = label_encoder.fit_transform(featured_columns['service'])
featured_columns['flag'] = label_encoder.fit_transform(featured_columns['flag'])
```

```
predicted_class = dataset[['class']]
predicted_class['class'] = label_encoder.fit_transform(predicted_class['class'])

predicted_class.head()
```

```
x = featured_columns[['protocol_type', 'service', 'flag', 'count', 'srv_count']]
y = predicted_class['class']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.9, random_state = 5)

print(x)
print(y)
```

```
# Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier

random_forest_model = RandomForestClassifier()
random_forest_model.fit(x_train, y_train.ravel())
y_pred = random_forest_model.predict(x_test)


print("Accuracy - Random Forest Classifier:\n{0:.3f}".format(metrics.accuracy_score(y_test, y_pred)))

cm_rfc = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n",cm_rfc)

print(classification_report(y_test, y_pred))
```

```
# Naive Bayes Classifier

from sklearn.naive_bayes import GaussianNB

naive_bayes_model = GaussianNB()
naive_bayes_model.fit(x_train, y_train.ravel())
y_pred = naive_bayes_model.predict(x_test)
print("Accuracy - Naive Bayes Classifier:\n{0:.3f}".format(metrics.accuracy_score(y_test, y_pred)))
cm_nbc = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n",cm_nbc)
print(classification_report(y_test, y_pred))
```

```
#Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier

decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(x_train, y_train.ravel())
y_pred = decision_tree_model.predict(x_test)
print("Accuracy - Decision Tree Classifier:\n{0:.3f}".format(metrics.accuracy_score(y_test, y_pred)))
cm_dtc = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n",cm_dtc)
print(classification_report(y_test, y_pred))
```
